<h1 style = "color : dodgerblue"> Hierarchical Clustering </h1>

<h2 style = "color : DeepSkyBlue"> An Overview of Hierarchical Clustering </h2>

* Hierarchical clustering is an unsupervised machine-learning clustering strategy.

* Unlike K-means clustering, tree-like morphologies are used to bunch the dataset, and dendrograms are used to create the hierarchy of the clusters.

<h2 style = "color : DeepSkyBlue"> What is Hierarchical Clustering? </h2>

* Hierarchical clustering is a type of unsupervised learning method used to group similar objects into clusters.

* It builds a hierarchy of clusters that can be represented in a tree-like structure known as a dendrogram.

* The basic idea is to create a cluster tree, where each group (cluster) is connected to other groups through a tree, allowing you to see at which level or distance each point was merged into a cluster.

* Hierarchical clustering is a connectivity-based clustering model that groups the data points together that are close to each other based on the measure of similarity or distance. The assumption is that data points that are close to each other are more similar or related than data points that are farther apart.

* A dendrogram, a tree-like figure produced by hierarchical clustering, depicts the hierarchical relationships between groups. Individual data points are located at the bottom of the dendrogram, while the largest clusters, which include all the data points, are located at the top. In order to generate different numbers of clusters, the dendrogram can be sliced at various heights.

* The dendrogram is created by iteratively merging or splitting clusters based on a measure of similarity or distance between data points. Clusters are divided or merged repeatedly until all data points are contained within a single cluster, or until the predetermined number of clusters is attained.

* We can look at the dendrogram and measure the height at which the branches of the dendrogram form distinct clusters to calculate the ideal number of clusters. The dendrogram can be sliced at this height to determine the number of clusters.

<b style = "color : orangered"> Here, dendrograms are the tree-like morphologies of the dataset, in which the X axis of the dendrogram represents the features or columns of the dataset, and the Y axis of the dendrogram represents the Euclidian distance between data observations. </b>

<h2 style = "color : deepskyblue"> 1. Hierarchical Agglomerative Clustering </h2>

* It is also known as the bottom-up approach or hierarchical agglomerative clustering (HAC).

* A structure that is more informative than the unstructured set of clusters returned by flat clustering.

* This clustering algorithm does not require us to prespecify the number of clusters.

* Bottom-up algorithms treat each data as a singleton cluster at the outset and then successively agglomerate pairs of clusters until all clusters have been merged into a single cluster that contains all data. 

<h3 style = "color : royalblue"> Algorithm :  </h3>

given a dataset (d1, d2, d3, ....dN) of size N

<b>compute the distance matrix</b>

for i=1 to N:

   <b>as the distance matrix is symmetric about</b> 

   <b>the primary diagonal so we compute only lower</b>
   
   <b>part of the primary diagonal</b>
   
   for j=1 to i:
      
      dis_mat[i][j] = distance[di, dj] 

each data point is a singleton cluster

<b style = "color : orangered">repeat</b>
  
   merge the two cluster having minimum distance
   
   update the distance matrix

<b style = "color : orangered">until</b> only a single cluster remains

<h3 style = "color : royalblue"> Steps: </h3>


* Consider each alphabet as a single cluster and calculate the distance of one cluster from all the other clusters.

* In the second step, comparable clusters are merged together to form a single cluster. Lets say cluster (B) and cluster (C) are very similar to each other therefore we merge them in the second step similarly to cluster (D) and (E) and at last, we get the clusters [(A), (BC), (DE), (F)]

* We recalculate the proximity according to the algorithm and merge the two nearest clusters([(DE), (F)]) together to form new clusters as [(A), (BC), (DEF)]

* Repeating the same process; The clusters DEF and BC are comparable and merged together to form a new cluster. Were now left with clusters [(A), (BCDEF)].

* At last, the two remaining clusters are merged together to form a single cluster [(ABCDEF)].

<h3 style = "color : royalblue"> Python implementation of the above algorithm using the scikit-learn library:  </h3>

```python
from sklearn.cluster import AgglomerativeClustering 
import numpy as np 

# randomly chosen dataset 
X = np.array([[1, 2], [1, 4], [1, 0], 
			[4, 2], [4, 4], [4, 0]]) 

# here we need to mention the number of clusters 
# otherwise the result will be a single cluster 
# containing all the data 
clustering = AgglomerativeClustering(n_clusters = 2).fit(X) 

# print the class labels 
print(clustering.labels_) 
```

<h3 style = "color : royalblue"> Agglomerative Hierarchical Clustering (Bottom-Up Approach):  </h3>

* Starts with each data point as an individual cluster.

* At each step, the two closest clusters are merged into a single cluster.

* This process continues until all data points belong to a single cluster.

<h2 style = "color : DeepSkyBlue"> 2. Hierarchical Divisive clustering </h2>

* It is also known as a top-down approach.

* This algorithm also does not require to prespecify the number of clusters.

* Top-down clustering requires a method for splitting a cluster that contains the whole data and proceeds by splitting clusters recursively until individual data have been split into singleton clusters.

<h3 style = "color : royalblue"> Algorithm :  </h3>


given a dataset (d1, d2, d3, ....dN) of size N

at the top we have all data in one cluster

the cluster is split using a flat clustering method eg. K-Means etc

<b style = "color : orangered"> repeat </b>

choose the best cluster among all the clusters to split

split that cluster by the flat clustering algorithm

<b style = "color : orangered"> until </b> each data is in its own singleton cluster

* Starts with all data points in a single cluster.

* At each step, the largest cluster is divided into two smaller clusters.

* This process continues until each data point is its own cluster.

<h2 style = "color : DeepSkyBlue"> Computing Distance Matrix </h2>

<b style = "color : #40E0D0">While merging two clusters we check the distance between two every pair of clusters and merge the pair with the least distance/most similarity. But the question is how is that distance determined. </b>

There are different ways of defining Inter Cluster distance/similarity. Some of them are:

<b style = "color : orangered">1. Min Distance:</b> Find the minimum distance between any two points of the cluster.

<b style = "color : orangered">2. Max Distance:</b> Find the maximum distance between any two points of the cluster.

<b style = "color : orangered">3. Group Average:</b> Find the average distance between every two points of the clusters.

<b style = "color : orangered">4. Wards Method:</b> The similarity of two clusters is based on the increase in squared error when two clusters are merged.

<h3 style = "color : RoyalBlue"> Implementation </h3>

```python
import numpy as np 
from scipy.cluster.hierarchy import dendrogram, linkage 
import matplotlib.pyplot as plt 

# randomly chosen dataset 
X = np.array([[1, 2], [1, 4], [1, 0], 
			[4, 2], [4, 4], [4, 0]]) 

# Perform hierarchical clustering 
Z = linkage(X, 'ward') 

# Plot dendrogram 
dendrogram(Z) 

plt.title('Hierarchical Clustering Dendrogram') 
plt.xlabel('Data point') 
plt.ylabel('Distance') 
plt.show()
```

<h2 style = "color : DeepSkyBlue"> Hierarchical Agglomerative vs Divisive Clustering </h2>

* Divisive clustering is more complex as compared to agglomerative clustering, as in the case of divisive clustering we need a flat clustering method as "subroutine" to split each cluster until we have each data having its own singleton cluster.

* Divisive clustering is more efficient if we do not generate a complete hierarchy all the way down to individual data leaves. The time complexity of a naive agglomerative clustering is O(n3) because we exhaustively scan the N x N matrix dist_mat for the lowest distance in each of N-1 iterations. Using priority queue data structure we can reduce this complexity to O(n2logn). By using some more optimizations it can be brought down to O(n2). Whereas for divisive clustering given a fixed number of top levels, using an efficient flat algorithm like K-Means, divisive algorithms are linear in the number of patterns and clusters.

* A divisive algorithm is also more accurate. Agglomerative clustering makes decisions by considering the local patterns or neighbor points without initially taking into account the global distribution of data. These early decisions cannot be undone. whereas divisive clustering takes into consideration the global distribution of data when making top-level partitioning decisions.

<h3 style = "color : RoyalBlue"> Working of Hierarchical Clustering </h3>

<b style = "color : #FF6700"> 1. Distance Metric: </b>

* Choose a suitable distance metric (e.g., Euclidean distance, Manhattan distance) to measure the similarity or dissimilarity between data points.

<b style = "color : #FF6700"> 2. Initial Clusters: </b>

<ul>
    <b><li style = "color : cyan">Agglomerative:</li></b> Each data point is initially considered an individual cluster.
    <b><li style = "color : cyan">Divisive:</li></b> All data points start in a single cluster.
</ul>

<b style = "color : #FF6700"> 3. Cluster Merging or Splitting: </b>

<ol>
    <b><li style = "color : cyan">Agglomerative:</li></b> 
    <ul>
        <li>Calculate the distance between each pair of clusters.</li>
        <li>Merge the two closest clusters.</li>
        <li>Update the distance matrix to reflect the new cluster.</li>
    </ul>
    <b><li style = "color : cyan">Divisive:</li></b>
    <ul>
        <li>Find the data point that is farthest from the cluster centroid.</li>
        <li>Create a new cluster with this data point.</li>
        <li>Reassign remaining data points to the closest cluster.</li>
    </ul>
</ol>

<b style = "color : #FF6700"> 4. Dendrogram: </b>

* A dendrogram is a tree-like diagram that visualizes the hierarchical structure of the clusters.

* The height of the branches in the dendrogram represents the distance between the merged or split clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram

# Sample data
X = np.array([[1, 2], [2, 3], [3, 4], [5, 6], [7, 8], [8, 9]])

# Create a hierarchical clustering model
clustering = AgglomerativeClustering(n_clusters = 2, linkage = 'single')

# Fit the model to the data
clustering.fit(X)

# Visualize the dendrogram
linked = linkage(X, 'single')
dendrogram(linked)
plt.show()

# Print the cluster labels
print(clustering.labels_)
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import KMeans

def divisive_clustering(data, k):
    """
    Performs divisive hierarchical clustering.

    Args:
        data: The input data.
        k: The desired number of clusters.

    Returns:
        A list of cluster labels for each data point.
    """

    # Initially, all data points are in a single cluster
    labels = np.zeros(len(data))

    while len(np.unique(labels)) > k:
        # Find the largest cluster
        unique_labels, counts = np.unique(labels, return_counts = True)
        largest_cluster_label = unique_labels[np.argmax(counts)]

        # Apply K-Means to split the largest cluster
        largest_cluster_indices = np.where(labels == largest_cluster_label)[0]
        largest_cluster_data = data[largest_cluster_indices]

        kmeans = KMeans(n_clusters = 2, random_state = 0).fit(largest_cluster_data)
        new_labels = kmeans.labels_

        # Assign new labels to the original data
        labels[largest_cluster_indices] = new_labels + np.max(labels) + 1

    return labels

# Example usage
# Sample data
X = np.array([[1, 2], [2, 3], [3, 4], [5, 6], [7, 8], [8, 9]])

# Desired number of clusters
k = 2

# Perform divisive clustering
labels = divisive_clustering(X, k)

# Visualize the dendrogram
linked = linkage(X, 'single')
dendrogram(linked)
plt.show()

print(labels)
```
