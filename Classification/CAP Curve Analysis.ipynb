<h1 style = "color : DarkTurquoise"> CAP (Cumulative Accuracy Profile) Curve </h1>

<h2 style = "color : cyan"> What is CAP curve? </h2>

* A CAP (Cumulative Accuracy Profile) curve is a performance measurement for classification problems.

* It is used to evaluate a model by comparing the current curve to both the 'perfect' or 'ideal' curve and a 'randomized' curve.

* A decent or good model will have a CAP that is in the middle of the perfect and random curves.

* The closer a model is to the perfect CAP, the better is.



<b style = "font-size : 25px">CAP Curve</b>

<h3 style = "color : orangered; font-size : 25px">There are two methods to analyze the performance using CAP curve:</h3>

<b style = "color : Turquoise; font-size : 15px">Area Under Curve:</b> We calculate the Accuracy Rate(AR) by calculating area under the perfect model and the random model (aP), and calculating area under the prediction model and random model (aR).

Accuracy Rate (AR) = aR / aP

Higher the AR, that is closer to the 1, better is the model.
