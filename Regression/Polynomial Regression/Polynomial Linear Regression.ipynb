<h1 style = "color : dodgerblue"> Polynomial Linear Regression </h1>

<h2 style = "color : MediumAquamarine"> Why Polynomial Regression? </h2>

* A simple linear regression algorithm only works when the relationship between the data is linear. 

* But suppose we have non-linear data, then linear regression will not be able to draw a best-fit line. Simple regression analysis fails in such conditions. 

* Consider the below diagram, which has a non-linear relationship, and you can see the linear regression results on it, which does not perform well, meaning it does not come close to reality. 

* Hence, we introduce it to overcome this problem, which helps identify the curvilinear relationship between independent and dependent variables.

<h2 style = "color : MediumAquamarine"> What is a Polynomial Regression? </h2>

* Polynomial regression is a form of Linear regression where only due to the Non-linear relationship between dependent and independent variables, we add some polynomial terms to linear regression to convert it into Polynomial Regression in Machine Learning.

* The relationship between the dependent variable and the independent variable is modeled as an nth-degree polynomial function. When the polynomial is of degree 2, it is called a quadratic model; when the degree of a polynomial is 3, it is called a cubic model, and so on.

* Suppose we have a dataset where variable X represents the Independent data and Y is the dependent data. Before feeding data to a mode in the preprocessing stage, we convert the input variables into polynomial terms using some degree.

* Consider an example my input value is 35, and the degree of a polynomial is 2, so I will find 35 power 0, 35 power 1, and 35 power 2 this helps to interpret the non-linear relationship in data.

* The equation of polynomials becomes something like this.



* The degree of order which to use is a Hyperparameter, and we need to choose it wisely. But using a high degree of polynomial tries to overfit the data, and for smaller values of degree, the model tries to underfit, so we need to find the optimum value of a degree.

* Polynomial Regression in Machine Learning models are usually fitted with the method of least squares. The least square method minimizes the variance of the coefficients under the Gauss-Markov Theorem.

<h2 style = "color : Turquoise"> Importing Libraries </h2>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h2 style = "color : Turquoise"> Importing Data Set </h2>

```python
df = pd.read_excel('RBC in Human.xlsx')
```

```python
# top 5 records
df.head()
```

```python
# bottom 5 records
df.tail()
```

```python
# columns : 2
# rows : 26
df.shape
```

```python
# Splitting dataset into X and Y
# X : Age
# Y : RBC
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(Y)
```

<h2 style = "color : Turquoise"> Training the Linear Regression model on the whole dataset </h2>

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, Y)
```

<h2 style = "color : Turquoise"> Training the Polynomial Regression model on the whole dataset </h2>

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 6)
X_poly = poly_reg.fit_transform(X)

linear_reg_2 = LinearRegression()
linear_reg_2.fit(X_poly, Y)
```

<h2 style = "color : Turquoise"> Visualising the Linear Regression results </h2>

```python
plt.scatter(X, Y, color = 'red')
plt.plot(X, linear_reg.predict(X), color = 'blue')
plt.title('Grade Vs Salary - Linear Regression')
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.show()
```

<h2 style = "color : Turquoise"> Visualising the Polynomial Regression results </h2>

```python
plt.scatter(X, Y, color = 'red')
plt.plot(X, linear_reg_2.predict(X_poly), color = 'blue')
plt.title('Grade Vs Salary - Polynomial Regression')
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.show()
```

<h2 style = "color : Turquoise"> Predicting a new result with Linear Regression </h2>

```python
linear_reg.predict([[11]])
```

<h2 style = "color : Turquoise"> Predicting a new result with Polynomial Regression </h2>

```python
linear_reg_2.predict(poly_reg.fit_transform([[12.1]]))
```
