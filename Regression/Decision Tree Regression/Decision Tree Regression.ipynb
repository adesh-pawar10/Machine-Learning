<h2 style = "color : MediumAquamarine"> Decision Tree Regression </h2>

* Decision tree regression observes features of an object and trains a model in the structure of a tree to predict data in the future to produce meaningful continuous output. 

* Continuous output means that the output/result is not discrete, i.e., it is not represented just by a discrete, known set of numbers or values.

<b style = "color : orangered">Discrete output example:</b> A weather prediction model that predicts whether or not therell be rain on a particular day. 

<b style = "color : orangered">Continuous output example:</b> A profit prediction model that states the probable profit that can be generated from the sale of a product.

<h2 style = "color : MediumAquamarine"> CART(Classification And Regression Tree) for Decision Tree </h2>

* CART is a predictive algorithm used in Machine learning and it explains how the target variables values can be predicted based on other matters.

* It is a decision tree where each fork is split into a predictor variable and each node has a prediction for the target variable at the end.

* The term CART serves as a generic term for the following categories of decision trees:

    1. Classification Trees: The tree is used to determine which class the target variable is most likely to fall into when it is continuous.

    2. Regression trees: These are used to predict a continuous variables value. 

<h2 style = "color : MediumAquamarine"> CART Algorithm </h2>

<b style = "color : SteelBlue; font-size: 110%">Classification and Regression Trees (CART) is a decision tree algorithm that is used for both classification and regression tasks. It is a supervised learning algorithm that learns from labelled data to predict unseen data.</b>

* <b style = "color : orangered; font-size : 125%">Tree structure:</b> CART builds a tree-like structure consisting of nodes and branches. The nodes represent different decision points, and the branches represent the possible outcomes of those decisions. The leaf nodes in the tree contain a predicted class label or value for the target variable.

* <b style = "color : orangered; font-size : 125%">Splitting criteria:</b> CART uses a greedy approach to split the data at each node. It evaluates all possible splits and selects the one that best reduces the impurity of the resulting subsets. For classification tasks, CART uses Gini impurity as the splitting criterion. The lower the Gini impurity, the more pure the subset is. For regression tasks, CART uses residual reduction as the splitting criterion. The lower the residual reduction, the better the fit of the model to the data.

* <b style = "color : orangered; font-size : 125%">Pruning:</b> To prevent overfitting of the data, pruning is a technique used to remove the nodes that contribute little to the model accuracy. Cost complexity pruning and information gain pruning are two popular pruning techniques. Cost complexity pruning involves calculating the cost of each node and removing nodes that have a negative cost. Information gain pruning involves calculating the information gain of each node and removing nodes that have a low information gain.

<h2 style = "color : MediumAquamarine"> How does CART algorithm works? </h2>

The CART algorithm works via the following process:

* The best-split point of each input is obtained. 
* Based on the best-split points of each input in Step 1, the new best split point is identified. 
* Split the chosen input according to the best split point. 
* Continue splitting until a stopping rule is satisfied or no further desirable splitting is available.

CART algorithm uses Gini Impurity to split the dataset into a decision tree .It does that by searching for the best homogeneity for the sub nodes, with the help of the Gini index criterion.

<h2 style = "color : Turquoise"> Importing Libraries </h2>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h2 style = "color : Turquoise"> Importing Data Set </h2>

```python
df = pd.read_excel('Salary_Data_Decision_Tree_Regression.xlsx')
```

```python
# top 5 records
df.head()
```

```python
# bottom 5 records
df.tail()
```

```python
# columns : 3
# rows : 12
df.shape
```

```python
# Splitting dataset into X and Y
# X : Grade
# Y : Salary
X = df.iloc[:, 1:-1].values
Y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(Y)
```

<h2 style = "color : Turquoise"> Training the dataset </h2>

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X, Y)
```

<h2 style = "color : Turquoise"> Predicting a new result </h2>

```python
regressor.predict([[5.5]])
```

<h2 style = "color : Turquoise"> Visualising the Decision Tree Regression </h2>

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color = 'orange')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Decision Tree Regression')
plt.xlabel('Grade')
plt.ylabel('Salary')
plt.show()
```
