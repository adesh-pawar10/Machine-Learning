```python
transactions = []

for i in range(0, 7466):
  transactions.append([str(df.values[i, j]) for j in range(0, 20)])
```

```python
transactions
```

<h2 style = "color : DeepSkyBlue"> Eclat Training on Dataset </h2>

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

<h2 style = "color : DeepSkyBlue"> Visualizing </h2>

<h3 style = "color : Royalblue"> Raw Results </h3>

```python
results = list(rules)
```

```python
results
```

```python
print(results)
```

<h2 style = "color : DeepSkyBlue"> Proper Display </h2>

```python
def inspect(results):
    movie_1         = [tuple(result[2][0][0])[0] for result in results]
    movie_2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(movie_1, movie_2, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Movie 1', 'Movie 2', 'Support'])
```
