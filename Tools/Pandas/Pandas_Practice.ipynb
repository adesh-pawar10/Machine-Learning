<h1 style = "color : salmon"> What is Pandas? </h1>

- Pandas is a powerful and open-source Python library. 

- The Pandas library is used for data manipulation and analysis. 

- Pandas consist of data structures and functions to perform efficient operations on data.

- Pandas is a powerful and versatile library that simplifies the tasks of data manipulation in Python.

- Pandas is well-suited for working with tabular data, such as spreadsheets or SQL tables.

- The Pandas library is an essential tool for data analysts, scientists, and engineers working with structured data in Python.

<b style = "font-size: 15px" > Here is a list of things that we can do using Pandas:</b>

- Data set cleaning, merging, and joining.
- Easy handling of missing data (represented as NaN) in floating point as well as non-floating point data.
- Columns can be inserted and deleted from DataFrame and higher-dimensional objects.
- Powerful group by functionality for performing split-apply-combine operations on data sets.
- Data Visualization.

<h2 style = "color : tomato"> Handling Missing Data </h2>

```python
import numpy as np
```

```python
data_frame.head()
```

```python
data_frame.describe()
```

```python
data_frame.iloc[2:3, 2:4] = np.nan
```

```python
data_frame.head()
```

```python
data_frame.describe()
```

```python
data_frame.isnull().sum()
```

```python
# drop the NaN values
# rows also removed in this method
data_frame.dropna(inplace = True)
data_frame.head()
```

```python
data_frame.reset_index(drop = True, inplace = True)
data_frame.head()
```

```python
data_frame.iloc[2:3, 2:4] = np.nan
data_frame.head()
```

```python
# finding mean/median/mode or most occuring value
# finding mean/median/mode or most occuring value for setosa

data_frame['petal_length'].fillna(data_frame['petal_length'].mean(), inplace = True)
```

```python
data_frame.head()
```

```python
# temp dataframe
temp = data_frame[data_frame['species'] == 'setosa']
temp.head()
```

```python
data_frame['petal_width'].fillna(temp['petal_width'].mean(), inplace = True)
data_frame.head()
```
