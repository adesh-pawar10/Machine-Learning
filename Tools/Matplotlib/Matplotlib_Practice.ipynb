<h1 style = "color : dodgerblue"> Matplotlib </h1>

Matplotlib is a powerful plotting library in Python used for creating static, animated, and interactive visualizations.

Matplotlibâ€™s primary purpose is to provide users with the tools and functionality to represent data graphically, making it easier to analyze and understand.

It was originally developed by John D. Hunter in 2003 and is now maintained by a large community of developers.

<h2 style = "color : dodgerblue"> Key Features of Matplotlib: </h2>


1. <b style = "color : #FF7F50">Versatility:</b> Matplotlib can generate a wide range of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more.

2. <b style = "color : #FF7F50">Customization:</b> It offers extensive customization options to control every aspect of the plot, such as line styles, colors, markers, labels, and annotations.

3. <b style = "color : #FF7F50">Integration with NumPy:</b> Matplotlib integrates seamlessly with NumPy, making it easy to plot data arrays directly.

4. <b style = "color : #FF7F50">Publication Quality:</b> Matplotlib produces high-quality plots suitable for publication with fine-grained control over the plot aesthetics.

5. <b style = "color : #FF7F50">Extensible:</b> Matplotlib is highly extensible, with a large ecosystem of add-on toolkits and extensions like Seaborn, Pandas plotting functions, and Basemap for geographical plotting.

6. <b style = "color : #FF7F50">Cross-Platform:</b> It is platform-independent and can run on various operating systems, including Windows, macOS, and Linux.

7. <b style = "color : #FF7F50">Interactive Plots:</b> Matplotlib supports interactive plotting through the use of widgets and event handling, enabling users to explore data dynamically.

<h2 style = "color : dodgerblue"> Import Library </h2>

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.scatter(x1, y1)
plt.scatter(x2, y2)

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.plot(x1, y1, 'o--')
plt.plot(x2, y2, '*--')
plt.show()
```

```python
import numpy as np
```

```python
x3 = np.array([1, 2, 3, 4, 5])
y3 = 2*(x3**4) + 100

plt.plot(x3, y3)
plt.show()
```

```python
x4 = np.arange(0, 10, 0.1)
y4 = 2*(x4**4) + 100

plt.plot(x4, y4)
plt.show()
```

<h2 style = "color : dodgerblue"> Matplotlib Part 2 </h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.25)
print(x)
print('-'*100)
y = x**4
print(y)
print('-'*100)

plt.plot(x, y, color = 'blue', marker = 'o', linestyle = 'dashed', linewidth = 2, markersize = 5)
plt.show()
```

```python
x1 = np.arange(0, 10, 0.25)
y1 = x1**4
y2 = x1**3

plt.plot(x1, y1, color = 'cyan', label = 'x^4', linewidth = 2, markersize = 5)
plt.plot(x1, y2, color = 'orange', label = 'x^3', linewidth = 2, markersize = 5)
plt.ylabel('Y', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.legend()
plt.title('Basic Chart')
plt.axis([0, 20, 0, 10000])
plt.grid()
plt.text(10, 8000, 'Chart is so good!', fontsize = 15)
plt.show()
```
